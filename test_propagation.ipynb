```python
import numpy as np
from propagation import sum_gaussians, prod_gaussians
from propagation import Gaussian
from propagation import Gaussian_deriv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = Gaussian()
y = Gaussian()
x.m = 0
x.v = 1
y.m = 0
y.v = 1

N = 100000
xs = np.random.normal(loc=x.m, scale=np.sqrt(x.v), size=N)
ys = np.random.normal(loc=y.m, scale=np.sqrt(y.v), size=N)

zs_correct = xs + ys
```

```python
res, grad = sum_gaussians(x, y)
zs_estimated = np.random.normal(loc=res.m, scale=np.sqrt(res.v), size=N)
```

```python
sns.distplot(zs_correct)
sns.distplot(zs_estimated)
plt.show()
```

```python
zs_prod_correct = xs * ys
```

```python
res = prod_gaussians(x, y)
zs_prod_estimated = np.random.normal(loc=res.m, scale=np.sqrt(res.v), size=N)

zs_prod_est_scaled = zs_prod_estimated * (1 / np.sqrt(2 * np.pi * res.v)) * (2 * np.pi * np.sqrt(x.m * y.m)) * \
    (1 / np.exp(-((x.m - y.m)**2) / (2 * (x.v + y.v))))
```

```python
sns.distplot(zs_prod_correct)
sns.distplot(zs_prod_estimated)
sns.distplot(zs_prod_est_scaled)
plt.show()
```
